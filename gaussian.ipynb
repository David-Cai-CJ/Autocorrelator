```python
import numpy as np
import matplotlib.pylab as plt
import scipy

def gau(x, x0=0, s=1):
    return 1/np.sqrt(2*np.pi)/s * np.exp(-(x-x0)**2/2/s**2)
```

```python
xx = np.linspace(-100,100,100)
yy = gau(xx, s = 20)
```

```python
np.sum(np.fft.fftshift(np.fft.ifft(np.fft.fft(yy)*np.fft.fft(yy)))*np.diff(xx).mean())
```

```python
# plt.plot(xx,yy)
plt.plot(xx, np.fft.fftshift(np.fft.ifft(np.fft.fft(yy)*np.fft.fft(yy))), label = 'conv')
plt.plot(xx, gau(xx,s=20*np.sqrt(2)) /np.sqrt(2), label = 'ana')
plt.legend()
```
